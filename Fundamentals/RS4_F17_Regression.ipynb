## Regression Analysis

연속형 변수 간 관계를 모델링해 적합도를 측정하는 분석 방법  
독립변수(Independent Variable)과 종속변수(Dependent Variable) 사이의 상호 관련성을 규명   

- 부모의 키 & 자식의 키 관계  
- 자동차의 스펙을 이용한 가격 예측  
- 1인당 국민 총소득 & 배기가스 배출량 사이 관계

회귀 개념의 시작  
- 영국 우생학자 F. Galton(1822-1911)  
- 아버지의 키 x와 자식의 키 y 사이의 직선 형태 기울기 < 1
- 세대와 무관하게 자식의 키는 전체 (통계적으로) 전체 평균에 수렴(회귀)하게 된다.  

오늘날의 회귀 = 두 개 이상의 변수 이상의 함수관계를 추구하는 통계적 방법   

선형회귀 = 두 변수 사이의 관계를 직선 형태로 가정하고 분석한다.  

___
**4가지 기본 가정**  
1. 선형성 - 종속변수 y와 독립변수 x 간 선형성 만족하는 특성
2. 독립성 - 다중 회귀분석에 필요한 기본 가정. 독립변수 x 간 상관관계 없이 독립성 만족
3. 등분산성 - 특정한 패턴 없이 고르게 분포했다. (등분산성)
4. 정규성 - 잔차가 정규성을 만족한다. 정규분포를 이루는지 확인한다.

___

### 지도학습의 두 종류: 분류와 회귀

분류 - 데이터 x의 feature 값을 활용해 데이터의 **클래스(범주형)** y를 추론  

회귀 - 데이터 x의 feature 값을 활용해 연관된 다른 데이터의 **정확한 값(연속형)** y를 추론

## Linear Regression

> $y = \beta \times x + c$ (parameters = $\beta$(회귀계수), $c$(오차))

목표: 주어진 데이터에 선형 식(formula)이 잘 맞도록 **회귀계수 및 오차를 구한다**

종속변수 y와 한 개 이상의 독립변수 x 간의 **선형 상관관계**를 모델링하는 회귀분석 기법  

### 머신러닝에서의 선형회귀모델 표기

> $H = W \times x + b$ ($H$: Hypothesis, $W$: weignt, $b$: bias)  

목표: n개의 점 데이터에 대해 잔차의 제곱의 합을 최소로 하는 $W$, $b$를 구하는 방법

대부분의 경우 $W$와 $b$가 단순 스칼라가 아닌 고차원의 행렬 형태.  
파라미터가 많을수록 모델 크기 & 학습 난이도 상승

## Logistic Regression (로지스틱 회귀분석)

데이터가 특정 범주에 속할 확률을 `0에서 1 사이 값으로 예측`,  
확률에 따라 가능성이 더 높은 범주에 속하는 것으로 `분류`하는 `지도학습` 알고리즘  
[0, 1] 사이의 확률분포에서 `특정 threshold를 기준으로 1과 0으로 이진분류`한다.
