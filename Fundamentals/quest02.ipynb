### np.linspace(시작 지점, 끝 지점, 개수)


```python
import numpy as np

x = np.linspace(0, 5, 30) # 0에서 10까지 동일한 간격의 수 100개를 생성
x
```

```python
y = np.sin(x)
z = np.cos(x)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5,3))

sns.lineplot(x=x, y=y)
sns.lineplot(x=x, y=z)

plt.show()
```

### All about Subplots

```python
# data
x = np.linspace(1, 100, 50)
y = np.random.randint(1, 50, size=50)


print(x.shape, y.shape)
```

```python
# Viz - 1행 2열의 도화지를 만들었습니다.
fig, axes = plt.subplots(1, 2, figsize=(6,3)) 

# matplot.pyplot 색연필로 그림 그리기
layer_1 = axes[0]
layer_1.set_title('Line graph drawn with matplot.pyplot', fontdict={'fontsize': 7})

layer_1.plot(x, y) # plot으로 선 그래프

# seaborn 싸인펜으로 그림 그리기
layer_2 = axes[1]
layer_2.set_title('Line graph drawn with seaborn', fontdict={'fontsize': 7})

sns.lineplot(x=x, y=y, ax=layer_2) # seabonr으로 선 그래프
```
