### 1. 라이브러리 불러오기

```python
import os
import cv2
import pixellib

import matplotlib.pyplot as plt
```

```python
def get_image_rgb(img_path):
    img_bgr = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

    return img_rgb
```

### 2. 인물사진 모드 구현


### 2.1. 3장 이상의 인물 모드 사진  

```python
fig, axes = plt.subplots(1, 3, figsize=(12,4))

plt.suptitle('Images for Portrait')

for i in range(3):
    pic_path = os.getcwd() + f"/image/img{i+1}.jpg"
    pic = get_image_rgb(pic_path)

    axes[i].imshow(pic)

plt.show()
```
